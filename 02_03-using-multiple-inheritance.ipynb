# Understanding multiple inheritance

```python
class A:
    def __init__(self):
        super().__init__()
        self.prop1 = "prop1"
        self.name = "Class A"
```

```python
class B:
    def __init__(self):
        super().__init__()
        self.prop2 = "prop2"
        self.name = "Class B"
```

```python
class C(B, A):
    def __init__(self):
        super().__init__()

    def showprops(self):
        print(self.prop1)
        print(self.prop2)
        print(self.name)
```

```python
# create the class and call showprops()
c = C()
print(C.__mro__)
c.showprops()
```
