```python
from dataclasses import dataclass
```

# Using data classes to represent data objects

```python
@dataclass
class Book:
    title: str
    author: str
    pages: int
    price: float

    # We can define methods in a dataclass like any other
    def bookinfo(self):
        return f"{self.title}, by {self.author}"
```

```python
# create some instances
b1 = Book("War and Peace", "Leo Tolstoy", 1225, 39.95)
b2 = Book("The Catcher in the Rye", "JD Salinger", 234, 29.95)
```

```python
# access fields
print(b1.title)
print(b2.author)
```

```python
# print the book itself - dataclasses provide a default implementation of the __repr__ function
print(b1)
```

```python
# comparing two dataclasses
b3 = Book("War and Peace", "Leo Tolstoy", 1225, 39.95)
print(b1 == b3)
```

```python
# change some fields, call a regular class method
b1.title = "Anna Karenina"
b1.pages = 864
print(b1.bookinfo())
```
