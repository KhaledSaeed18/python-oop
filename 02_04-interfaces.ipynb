```python
from abc import ABC, abstractmethod
```

# Using Abstract Base Classes to implement interfaces

```python
class GraphicShape(ABC):
    def __init__(self):
        super().__init__()

    @abstractmethod
    def calcArea(self):
        pass
```

```python
class JSONify(ABC):
    @abstractmethod
    def toJSON(self):
        pass
```

```python
class Circle(GraphicShape, JSONify):
    def __init__(self, radius):
        self.radius = radius

    def calcArea(self):
        return 3.14 * (self.radius ** 2)

    def toJSON(self):
        return f"{{ \"square\": {str(self.calcArea())} }}"
```

```python
c = Circle(10)
print(c.calcArea())
print(c.toJSON())
```
